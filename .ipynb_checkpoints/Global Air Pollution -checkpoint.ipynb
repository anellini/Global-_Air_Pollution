```python
#Displot by features
col=[ 'AQI Value', 'CO AQI Value',
       'Ozone AQI Value','NO2 AQI Value', 'PM2.5 AQI Value']
fig = plt.figure(figsize=(15,5))

for i in range(len(col)):
    plt.subplot(1,5,i+1)
    plt.title(col[i])
    sns.distplot(data,x=data[col[i]])
plt.tight_layout()
plt.show()
```

```python
#Boxplot by features and categories
col=[ 'AQI Value', 'CO AQI Value',
       'Ozone AQI Value','NO2 AQI Value', 'PM2.5 AQI Value']
fig = plt.figure(figsize=(15,10))

for i in range(len(col)):
    plt.subplot(2,3,i+1)
    plt.title(col[i])
    sns.boxplot(data=data,x=data[col[i]],y=data['AQI Category'])
plt.tight_layout()
plt.show()
```

```python
# Most Country have worst overall AQI value
overall = data[['Country','City','AQI Value','AQI Category']]
aql_haz = overall[overall['AQI Category'] == 'Hazardous']
haz = aql_haz.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value',ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=haz)
plt.title('Most Country have worst overall AQI value')
```

```python
# Most Country have good overall AQI value
overall_good = overall[overall['AQI Category'] == 'Good']
good = overall_good.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value').head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=good)
plt.title('Most Country have good overall AQI value')
```

```python
# Correlation heatmap
sns.heatmap(data.corr(),annot=True)
```
